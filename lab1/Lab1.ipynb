```python
import matplotlib.pyplot as plt  #import required libs
import numpy as np
import cv2


def extract_channels(image):
    img=cv2.imread(image)  #read image

    print img.shape
    red=img[:,:,0]
    green=img[:,:,1]#Extract channels simple numpy slicing
    blue=img[:,:,2]
    plt.imshow(img.astype('uint8'))  #convert to uint8 to show
    plt.show()
    plt.imshow(blue.astype('uint8'))
    plt.show()
    plt.imshow(red.astype('uint8'))
    plt.show()
    plt.imshow(green.astype('uint8'))
    plt.show()
 
    
extract_channels("images.jpg")
```

```python
import cv2 
def resize(image):
    img=cv2.imread(image)
    small=cv2.resize(img,(100,100),interpolation=cv2.INTER_LINEAR)  #opencv resizing 
    cv2.imwrite("small.bmp",small)
    plt.imshow(small)
    plt.show()
resize("images.jpg")
```
