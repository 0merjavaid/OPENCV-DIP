```python
# Name Muhammad Umer Javaid
# bese 5a
# 111238

#Task 1

import cv2,numpy as np
import matplotlib.pyplot as plt
im=cv2.imread("cartoon.tif",0)
im1=cv2.normalize(im.astype('float'),None,0.0,1.0,cv2.NORM_MINMAX)

def high_boost(img,A):
    ker=np.array([[0,-1,0],[-1,A+4,-1],[0,-1,0]])
    im= cv2.filter2D(img, -1, ker)
    im[im>1]=1
    im[im<0]=0
    return im
    
im2 = high_boost(im1,1) 
im3=(im1+im2)
im3[im3>1]=1
im3[im3<0]=0

im2 = high_boost(im1,3) 
im4=(im1+im2)
im4[im4>1]=1
im4[im4<0]=0

im3=cv2.normalize(im3.astype('float'),None,0,255,cv2.NORM_MINMAX).astype('uint8')
 

plt.subplot(311)
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.title("Original")
plt.show()
 
plt.subplot(312)
plt.imshow(im3, cmap='gray')
plt.axis('off')
plt.title("High boost A=1")

plt.show()
plt.subplot(313)
plt.imshow(im4, cmap='gray')
plt.axis('off')
plt.title("High boost A=3")



plt.show()

#And from experiment its visible that highboost with A=3 is like image*3
```

```python
#Task 3
import matplotlib.pyplot as plt
def Unsharp_masking(img):
    
    im=cv2.GaussianBlur(img,(199,199),0) 
    return img-im
img=cv2.imread("pears.jpg",0)
im=Unsharp_masking(img)

plt.subplot(211)
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.title("vertical")
plt.show()
 

```

```python

def gradient(img,flag):
    if(flag):
        i=img.copy()
        img=np.roll(img,1)
    else:
        i=img.copy()
        img=np.roll(img,1,0)
    return img-i

def task4():
    cartoon = cv2.imread("cartoon.tif", 0)


    vertical=gradient(cartoon,0)
    hor=gradient(cartoon,1)

    
    plt.subplot(211)
    plt.imshow(vertical, cmap='gray')
    plt.axis('off')
    plt.title("vertical")

    plt.subplot(212)
    plt.imshow(hor, cmap='gray')
    plt.axis('off')
    plt.title("Horizontal")

    plt.show() 
task4()
```
