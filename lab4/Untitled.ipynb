```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# loading image
#img0 = cv2.imread('SanFrancisco.jpg',)
img0 = cv2.imread('image.jpg')

# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)

# convolute with proper kernels
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # y

pewit=filter("prewitt")
prewitt=cv2.filter2D(img,-1,pewit)

pewit=filter("robetrs")
rob=cv2.filter2D(img,-1,pewit)

kir=filter("kirsch")
kirsch=cv2.filter2D(img,-1,kir)
 

print 'ORIGINAL'
plt.imshow(img,cmap = 'gray') 
plt.show()
print 'LAPLACIAN'
plt.imshow(laplacian,cmap = 'gray')  
plt.show()
print "SOBELX"
plt.imshow(sobelx,cmap = 'gray') 
plt.show()
print "SOBELY"
plt.imshow(sobely,cmap = 'gray') 
plt.show() 
print "PREWITT"

plt.imshow(prewitt,cmap = 'gray') 
plt.show() 
print 'ROBERTS'
plt.imshow(rob,cmap = 'gray') 
plt.show() 
print "KIRSCH"

plt.imshow(kirsch,cmap = 'gray') 
plt.show() 
```
